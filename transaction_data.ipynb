```python
import pandas as pd
```

```python
transactions_data = pd.read_csv("data/bank_transactions.csv")
```

```python
print(list(transactions_data.columns))
```

```python
transaction = transactions_data.filter(['TransactionID', 'AccountID', 'TransactionAmount', 'TransactionDate', 'TransactionType',
                                       'Location','DeviceID', 'MerchantID', 'TransactionDuration', 'LoginAttempts'], axis=1)
transaction.to_csv("data/transaction.csv")
transaction.head()
```

```python
accounts = transactions_data.filter(['AccountID', 'CustomerAge', 'CustomerOccupation', 'AccountBalance','PreviousTransactionDate'],axis=1)
accounts.to_csv("data/accounts.csv", index = False)
accounts.head()
```

```python
devices = transactions_data.filter(['DeviceID', 'IP Address'],axis=1)
devices.to_csv("data/devices.csv", index = False)
devices.head()
```

```python
merchants = transactions_data.filter(['MerchantID', 'Channel'],axis=1)
merchants.to_csv("data/merchants.csv", index = False)
merchants.head() 
```
